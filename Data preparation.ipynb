## Load the complete data as a dataframe in pandas

```python
import pandas as pd

# Load CSV file into a Pandas DataFrame
all_data = pd.read_csv(r'C:\Users\msata\Desktop\Btech_project_2023\test script 2\all_data.csv', index_col=0)
```

```python
all_data.head() #See the top columns
```

The above output shows some of the Na values. The genes which have got more than 15% percent of NA values are to be removed from consideration [reference can be found here](https://www.sciencedirect.com/science/article/abs/pii/S1476927119300040?via%3Dihub)

```python
all_data.shape #see the dimensions
```

```python
all_data_copy = all_data.copy() # Store it as a copy to preserve original dataset 
```

```python
# Calculate the threshold for the number of non-null values required
thresh = int(0.85 * len(all_data.columns))
```

```python
# Drop rows (genes) with less than the threshold number of non-null values
all_data_copy = all_data_copy.dropna(thresh=thresh)
```

```python
all_data_copy.shape
```

As you can see $27898-23768=4130$ genes have been discarded

## Now separate the DEG's subset from the entire data

To delete all rows from all_data_copy whose names are not present in the ID column of degs_ids, you can use the **isin()** method to create a Boolean mask indicating which rows have matching IDs, and then use the **loc[]** accessor to select the matching rows from all_data_copy.

```python
# Create Boolean mask to select rows with matching IDs
mask = all_data_copy.index.isin(degs_ids['ID'])
```

```python
mask.sum() # Note that there were 6397 DEG's but we got 4 less as we removed if missing data was more than 15%
```

```python
# Select matching rows from all_data_copy
final_data = all_data_copy.loc[mask]
```

```python
final_data.head()
```

```python
final_data.describe()
```

```python
final_data.isna().any().count()
```

In the dataframe that we have above, the rows are corresponding to the genes and the columns are corresponding to sample ids. For doing any ML analysis it is a standard convection that the paramaeters/variables are maintained as columns. Through this we can use the pre-implemented ML libraries and modules. Therefore we will take a transpose of the above dataframe.

```python
final_df = final_data.transpose()
```

```python
final_df.describe()
```

## Handle the missing values using KNN Impute

```python
from sklearn.impute import KNNImputer
```

```python
# Create KNN imputer object
imputer = KNNImputer(n_neighbors=5)
```

```python
final_df.isna().sum().sum() # These are the number of NAs values
```

```python
# Fill missing values using KNN imputer
final_df = pd.DataFrame(imputer.fit_transform(final_df), columns=final_df.columns)
```

```python
final_df.isna().sum().sum() # These are the number of NAs values
```

```python
final_df.index
```

```python
final_data.columns
```

```python
# Set the old column names as row names
final_df = final_df.set_index(final_data.columns)
```

```python
final_df.index #These are the row names (also called as index in ML)
```
